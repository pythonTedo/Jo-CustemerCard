# Machine Learning

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV

from sklearn.linear_model import LogisticRegressionCV

from sklearn.svm import SVC

from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier

from sklearn.neighbors import KNeighborsClassifier

from sklearn.decomposition import PCA

from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, roc_auc_score, plot_confusion_matrix, plot_roc_curve

from lightgbm import LGBMClassifier
```

Before the cluserting I would like to assume a standard classification problem - predict the Bundesland based on the data we are giving. So for now Ill assume that the predictor will be B_Land.

After that I will threat the data as unsupervised in order to see how clustering methods will split and classify the data itself.

```python
df_dropped = df.drop(["KaufKraft_Lebensmittelhandel_KOPF", "KaufKraft_Drogeriefachhandel_KOPF", "Kaufkraft_KOPF", "ANZAHL_Verbrauchermärkte_im Umkreis von 20min", "ANZAHL_Diskonter_im Umkreis von 20min"], axis=1)
df_dropped
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(df_dropped.drop("B_LAND", axis=1), df_dropped["B_LAND"], train_size=0.8)
```

```python
scaler = MinMaxScaler()

X_train_sc = scaler.fit_transform(X_train)
X_test_sc = scaler.transform(X_test)
```

```python
models_to_fit = [LogisticRegressionCV(), RandomForestClassifier(), SVC(), AdaBoostClassifier(), LGBMClassifier()]
```

```python
def GetModelScores (estimator, X_train, y_train):
    scores_train = pd.DataFrame(columns= ['Accuracy','F1 Score','Precision','Recall','ROC_AUC'])
    scores_test = pd.DataFrame(columns= ['Accuracy','F1 Score','Precision','Recall','ROC_AUC'])
    
    model = estimator
    model.fit(X_train, y_train)
    
    prediction_train = model.predict(X_train)
    #prediction_test = model.predict(X_test)
    
    try:
        score_train = model.predict_proba(X_train)[:,1]
        roc_train= roc_auc_score(y_train, score_train, average = "weighted")
    except:
        roc_train = 0

    '''
    try:
        score_test = model.predict_proba(X_test)[:,1]
        roc_test= roc_auc_score(y_test, score_test, average = "weighted")
    except:
        roc_test = 0
    '''
    
    
    scores_train['Accuracy'] = accuracy_score(y_train, prediction_train)*100,
    scores_train['F1 Score'] = f1_score(y_train, prediction_train, average = "weighted")*100,
    scores_train['Precision'] = precision_score(y_train, prediction_train, average = "weighted")*100,
    scores_train['Recall'] = recall_score(y_train, prediction_train, average = "weighted")*100,
    scores_train['ROC_AUC'] = roc_train*100
    
    '''
    scores_test['Accuracy'] = accuracy_score(y_test, prediction_test)*100,
    scores_test['F1 Score'] = f1_score(y_test, prediction_test, average = "weighted")*100,
    scores_test['Precision'] = precision_score(y_test, prediction_test, average = "weighted")*100,
    scores_test['Recall'] = recall_score(y_test, prediction_test, average = "weighted")*100,
    scores_test['ROC_AUC'] = roc_test*100
    '''
    print(scores_train)
    
```

```python
for i in models_to_fit:
    print(f"Fitting model {i.__class__}")
    GetModelScores(i, X_train_sc, Y_train)
    print("=====================================================")
```

```python
lg = LogisticRegressionCV()
lg.fit(X_train_sc, Y_train)
```

```python
plot_confusion_matrix(lg, X_train_sc, Y_train)
```

```python
plot_confusion_matrix(lg, X_test_sc, Y_test)

f1_score(Y_test, lg.predict(X_test_sc), average = "weighted")
```

The thata clearly underfits.

Ways to tackle this problem:
1. More data
2. More complex model
3. Get more features which we can use

```python
k_fold = StratifiedKFold(n_splits = 5)
parameters = {
    "n_estimators": [50, 200, 400],
    "max_depth": [10, 50, 70]
}
grid_forest = GridSearchCV(RandomForestClassifier(), parameters, scoring = 'accuracy', cv = k_fold, n_jobs =-1)
grid_forest.fit(X_train_sc, Y_train)

grid_forest.best_estimator_
```

```python
grid_forest.cv_results_
```

```python

plot_confusion_matrix(grid_forest, X_test_sc, Y_test)
f1_score(Y_test, grid_forest.predict(X_test_sc), average="weighted")
```

I will use UMAP technique in order to vizualize the data into 2 dimension scpase. This technique has eseveral advantages: Preservation of Global Structure - takes into account all points not just nearby


```python
import umap
ump = umap.UMAP(random_state=42)
umap_data = ump.fit_transform(df.drop(["B_LAND", "ANZAHL_Diskonter_im Umkreis von 20min", "ANZAHL_Verbrauchermärkte_im Umkreis von 20min"], axis=1))
plt.figure(figsize=(20,12))
sns.scatterplot(x=umap_data[:,0], y=umap_data[:,1], hue=df["B_LAND"], palette='cool')
plt.show()
```
