```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("data/Filialdaten.xlsx", index_col="FILIALE")
```

```python
df.info()
```

# Load data and Data analysis

```python
df.head()
```

```python
gr_bland = df.groupby("B_LAND").agg("sum")
gr_bland
```

```python
df_grouped_anteils= df[["B_LAND","Anteil Clever", "Anteil Ja!Natürlich", "Anteil Feinkost", "Anteil Obst&Gemüse", "Anteil BILLA Marke", "Anteil BILLA Corso", "Anteil Getränke ohne Alkohol", "Anteil Alkohol"]].groupby("B_LAND").agg("sum")
for index, row in df_grouped_anteils.iterrows():
    plt.figure(figsize=(10, 8))  # Adjust the figure size as needed
    sns.barplot(row.index, row.values)
    plt.xlabel('Sum')
    plt.ylabel('Features')
    plt.title(f'B_LAND {index}')
    plt.xticks(rotation=45)
    plt.show()
```

The data is grouped by Bundesland. 
We can see that the majority of the discounts wich people use is the Feinkost and Fruits and Vegitables. 
The Alcohol was also strong 3rd position discount, especially at Bundesland 9, 8, 7 and 6, but pay attention to the scale... All of them alcohol exists as 3rd position.

We can aslo see strong correclation between Clever and Biulla because Clever is sub-company of BILLA?

This can help us to target based on the region and the revenue the discounts whcih can be offered to people.
